```python
import pandas as pd
```

```python
df = pd.read_csv("LMB_2024_db.csv", encoding='latin1')
df.head()
```

```python
Complete_data = df[df["PA"]>=36]
print(df.shape)
Complete_data.shape
```

```python
lw_1_1 = 0.536 + 0.116
lw_2_1 = 0.883 + 0.116
lw_3_1 = 1.230 + 0.116
lw_hr_1 = 1.521 + 0.116
lw_w_1 = 0.371 + 0.116
lw_hbp_1 = 0.007 + 0.116

Complete_data["1B"] = Complete_data["H"] - Complete_data["2B"] - Complete_data["3B"] - Complete_data["HR"]

x = (sum(Complete_data["1B"])*lw_1_1 + sum(Complete_data["2B"])*lw_2_1 + sum(Complete_data["3B"])*lw_3_1 + sum(Complete_data["HR"])*lw_hr_1 + sum(Complete_data["BB"])*lw_w_1 +
     sum(Complete_data["HBP"])*lw_hbp_1) / sum(Complete_data["PA"])

wOBA_scale = round(Complete_data["OBP"].mean(),4) / x

wOBA_scale
```

```python
lw_1 = 0.536 * wOBA_scale
lw_2 = 0.883 * wOBA_scale
lw_3 = 1.230 * wOBA_scale
lw_hr = 1.521 * wOBA_scale
lw_w = 0.371 * wOBA_scale
lw_hbp = 0.007 * wOBA_scale

```

```python
Complete_data ["wOBA"] = ((Complete_data["1B"]*lw_1) + (Complete_data["2B"]*lw_2) + (Complete_data["3B"]*lw_3) + (Complete_data["HR"]*lw_hr) + ((Complete_data["BB"] - Complete_data["IBB"])*lw_w) +
 (Complete_data["HBP"]*lw_hbp))/ (Complete_data["AB"] + Complete_data["BB"] - Complete_data["IBB"] + Complete_data["HBP"] + Complete_data["SF"] )
Complete_data.head()
```

```python
# Sort the DataFrame by 'OPS+' in descending order
Complete_data = Complete_data.sort_values(by=['wOBA'], ascending=False)
Complete_data.head()[["Name", "OBP", "wOBA"]]
```

```python
Complete_data["Age"] = Complete_data["Age"].astype(int)
young_stars = Complete_data[Complete_data["Age"]<=25]
young_stars = young_stars.sort_values(by=['wOBA'], ascending=False)
young_stars.head()[["Name", "OBP", "wOBA"]]
```

```python
l_wOBA = Complete_data["wOBA"].mean()
l_r_pa = sum(Complete_data["R"])/sum(Complete_data["PA"])
print(f'League wOBA = {l_wOBA}\n League_ run / PA {l_r_pa}')
```

```python
Complete_data["wRAA"] = ((Complete_data["wOBA"] - l_wOBA) / wOBA_scale) * Complete_data["PA"]
Complete_data["wRC"] = (((Complete_data["wOBA"] - l_wOBA) / wOBA_scale) + l_r_pa) * Complete_data["PA"]
l_wr_pa = sum(Complete_data["wRC"]) / sum(Complete_data["PA"])
Complete_data["wRC+"] = ((((Complete_data["wRAA"]/Complete_data["PA"]) + l_r_pa) + (l_r_pa - (Complete_data["PF"] * l_r_pa))) / l_wr_pa) * 100
Complete_data.head()
```

```python
young_stars = Complete_data[Complete_data["Age"]<=25]
young_stars = young_stars.sort_values(by=['wRC+'], ascending=False)
young_stars.head()[["Name", "OBP", "wOBA","wRAA","wRC","wRC+"]]
```
